```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['last'] == 'Doe'
```

```python
filt = (df['last'] == 'Doe')
```

```python
df[filt]
```

```python
df[df['last'] == 'Doe'] # another way to do the same thing
```

```python
df.loc[filt] # this is also giving the same result 
```

```python
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')
```

```python
(df['last'] == 'Doe') & (df['first'] == 'John')
```

```python
df.loc[~filt, 'email']
```
