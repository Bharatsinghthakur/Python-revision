```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['first'] + ' ' + df['last']
```

```python
df['full name'] = df['first'] + ' ' +df['last']
```

```python
df
```

```python
df.drop(columns=['first','last'])
```

```python
# if we want to remove these changes, then we can do 
df.drop(columns=['first','last'],inplace=True)
```

```python
df
```

```python
df['full name'].str.split(' ',expand=True)
```

```python
df[['first','last']] = df['full name'].str.split(' ',expand=True)
```

```python
df
```

```python
new_row = pd.DataFrame([{'first':'bharat'}])
```

```python
df = pd.concat([df,new_row],ignore_index=True)
```

```python
people = {
    "first": ["Jack", 'Neo', 'Niko'], 
    "last": ["Singh", 'Singh', 'Singh'], 
    "email": ["jacksingh@gmail.com", 'Neosingh@email.com', 'nikosingh@email.com']
}
df2 = pd.DataFrame(people)
```
